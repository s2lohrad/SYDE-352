The [Python Control Systems Library](https://python-control.readthedocs.io/en/0.8.3/index.html) is a Python package that implements basic operations for analysis and design of feedback control systems.

To install this library uese bellow command:

```
pip install control
```

Now that we have the control library installed, it should be imported:

```
import control as co
```

```
import numpy as np
import matplotlib.pyplot as plt
```

There are several ways to define the system's model. For example we can use tf function which gets the denominator and numerators' coefficients:
$$\frac{2 s + 3}{s^2 + 3 s + 5}$$

```
sys = co.tf([2,3],[1,3,5])
```

Here we define t for plotting porpous. Then use several different functions to plot step-response, root-locus, bode and nyquist diagrams of the mentioned transfer function:

```
t = np.linspace(1, 10, 1000)
```

```
t1, y1 = co.step_response(sys, t)

plt.plot(t1, y1)
plt.grid
```

```
co.bode_plot(sys, dB=True)
```

```
co.nyquist_plot(sys)
```

```
co.root_locus(sys)
```
